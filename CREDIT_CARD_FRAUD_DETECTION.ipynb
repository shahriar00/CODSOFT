**Label Encoder unsing for the gender labeling**

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df_train['gender'] = le.fit_transform(df_train['gender'])
```

```python
df_test['gender'] = le.fit_transform(df_test['gender'])
```

**All information check in one frame**

**Delete unnessery column**

```python
df_train.drop(columns = ['Unnamed: 0','cc_num','merchant','first','last','trans_num','unix_time','street','category','trans_date_trans_time','city','job','dob','state'],inplace = True)
```

```python
df_test.drop(columns = ['Unnamed: 0','cc_num','merchant','first','last','trans_num','unix_time','street','category','trans_date_trans_time','city','job','dob','state'],inplace = True)
```

```python
df_train.head(10)
```

**Null Value check**

```python
df_train.isnull().sum()
```

```python
X_train = df_train.drop(columns = 'is_fraud',axis = 1)
X_test = df_test.drop(columns = 'is_fraud',axis = 1)

Y_train = df_train['is_fraud']
Y_test = df_test['is_fraud']
```

**Data Spliting**

```python
x_train,x_test,y_train,y_test = train_test_split(X_train,Y_train,test_size = 0.2)
```

**Logistic Regression**

```python
logistic = LogisticRegression()
logistic.fit(x_train,y_train)
train_pred = logistic.predict(x_train)
accuracy = accuracy_score(train_pred,y_train)
print("Accuracy: ",accuracy)
```

```python
logistic.fit(x_test,y_test)
test_pred = logistic.predict(x_test)
accuracy = accuracy_score(test_pred,y_test)
print("Accuracy: ",accuracy)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

**Decision Tree**

```python
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
train_pred = dt.predict(x_train)
accuracy = accuracy_score(train_pred,y_train)
print("Accuracy: ",accuracy)
```

```python
dt.fit(x_test,y_test)
test_pred = dt.predict(x_test)
accuracy = accuracy_score(test_pred,y_test)
print("Accuracy: ",accuracy)
```

**Random Forest**

```python
rf = RandomForestClassifier()
rf.fit(x_train,y_train)
train_pred = rf.predict(x_train)
accuracy = accuracy_score(train_pred,y_train)
print("Accuracy: ",accuracy)
```

```python
rf.fit(x_test,y_test)
test_pred = rf.predict(x_test)
accuracy = accuracy_score(test_pred,y_test)
print("Accuracy: ",accuracy)
```
